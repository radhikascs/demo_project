```python
groupdata = pd.DataFrame(plot_df.groupby(['Registration_date','Medium'])['Medium'].count().fillna(0).reset_index(name ='Count'))
groupdata.dtypes
```

```python
dfp_pivot = pd.pivot_table(groupdata,values='Count',index ='Registration_date',columns='Medium',aggfunc=sum).fillna(0)

dfp_pivot = pd.DataFrame(dfp_pivot.to_records())
dfp_pivot
```

```python
from bokeh.models import ColumnDataSource, Select
from bokeh.io import show, output_notebook 
#Figure objects have many glyph methods that can be used to draw vectorized graphical glyphs. example of glyphs-circle, line, scattter etc. 
from bokeh.plotting import figure 
#To create intractive plot we need this to add callback method.
from bokeh.models import CustomJS 
#This is for creating layout
from bokeh.layouts import column, row
from bokeh.models import FactorRange
from bokeh.io import curdoc
output_notebook()

iactive_groupdata = groupdata.copy()
```

```python
iactive_groupdata['Registration_date'] = iactive_groupdata['Registration_date'].astype('str')
iactive_groupdata

filterdf1 = iactive_groupdata.loc[:,['Registration_date','Medium','Count']]
filterdf2 = filterdf1[filterdf1['Registration_date']=='2021-01-03']


Overall = ColumnDataSource(data=filterdf1)
Curr=ColumnDataSource({'x': filterdf1['Medium'], 'y': filterdf1['Count']})


#plot and the menu is linked with each other by this callback function
callback = CustomJS(args=dict(source=Overall, sc=Curr), code="""
var f = cb_obj.value;
sc.data['Medium']=[];
sc.data['Count']=[];
for(var i = 0; i <= source.get_length(); i++){
	if (source.data['Registration_date'][i] == f){
		sc.data['Medium'].push(source.data['Medium'][i])
		sc.data['Count'].push(source.data['Count'][i])
	 }
}   
   
sc.change.emit();
""")


vbar13=  iactive_groupdata.plot_bokeh.bar(
    x='Medium',
    y='Count',
    xlabel='Medium',
    ylabel='Count',
    title='Registration day wise',
    figsize =(600,400)
)

menu = Select(options=list(filterdf1['Registration_date'].unique()),value='2021-01-03', title = 'Registration-Date')  # drop down menu
menu.js_on_change('value', callback) # calling the function on change of selection
layout=column(menu, vbar13) # creating the layout
show(layout) # displaying the layout
curdoc().add_root(layout)
```
